```python
n= int(input("bạn muốn nhập vào bao nhiêu số ?"))
m=[]
for i in range(1,n+1):
  m.append(i)
print(m)
```

```python
def tong(m):
  a=sum(m)
  print("tổng các số tự nhiên của n là",a)
tong(m)
```

```python
def nhan(m):
    result = 1
    for x in m:
        result = result * x
    return result
print(nhan(m))
```
